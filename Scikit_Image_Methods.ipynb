## 1- RGB to grayscale Conversion (Image Transformation / Color Space Conversion)


```python
# Figure 1

from skimage import io, color
import matplotlib.pyplot as plt

# Get the image from GitHub
Image_url_1 = "https://raw.githubusercontent.com/Reem-KR/GEOL-558_Scikit-Image/main/Red_Sand_Beach.png"

# Load the RGB image
Image_1 = io.imread(Image_url_1)

# Remove the alpha channel if it exists (RGBA â†’ RGB)
if Image_1.shape[-1] == 4:
    Image_1 = Image_1[:, :, :3]  # Keep only the first three channels (RGB)

# Convert to Grayscale
gray_image = color.rgb2gray(Image_1)

# Display both images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Show original RGB image
ax[0].imshow(Image_1)
ax[0].set_title("Original Image of Red Sand Beach")
ax[0].axis("off")

# Show grayscale image
ax[1].imshow(gray_image, cmap='gray')
ax[1].set_title("Grayscale Image of Red Sand Beach")
ax[1].axis("off")

# Display both images
plt.show()
```

**Figure 1:**

This is an image I took visiting the Red Sand Beach in Maui, Hawaii. I chose this example to demonstrate how an image can be converted from color (RGB) to grayscale. This is a common image preprocessing technique used in computer vision tasks, as it simplifies the data by reducing it to a single intensity channel while still preserving the structural information. By implementing this code, I learned how to load an image from an external URL using skimage.io.imread(),handle images with an alpha channel (RGBA) by removing it and retaining just the RGB channels, and convert an RGB image to grayscale using skimage.color.rgb2gray().

## 2- Canny Edge Detection (Feature / Edge Detection)



```python
# Figure 2

from skimage import io, feature, color
import matplotlib.pyplot as plt

# Get the image from GitHub
Image_url_2 = "https://raw.githubusercontent.com/Reem-KR/GEOL-558_Scikit-Image/main/Cracks_at_Sunrise-on-Sea%2C_Eastern_Cape.jpg"

# Load image
Image_2 = io.imread(Image_url_2)

# Convert to grayscale if image is in RGB
if Image_2.ndim == 3:
    Image_2 = color.rgb2gray(Image_2)

# Apply Canny Edge Detection
edges = feature.canny(Image_2, sigma=1)

# Create a figure with two subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Show original grayscale image
ax[0].imshow(Image_2, cmap='gray')
ax[0].set_title("Original Image: Cracks at Sunrise-on-Sea")
ax[0].axis("off")

# Show Canny Edge-Detected image
ax[1].imshow(edges, cmap='gray')
ax[1].set_title("Canny Edge Detection")
ax[1].axis("off")

# Display both images
plt.show()
```

**Figure 2:**

 This is an image I found online of cracks or fractures in a rock that shows this circular flower-shaped pattern that I thought was cool. I chose this example to see what the Canny Edge Detection algorithm can detect in images since I read that this is a widely used technique in computer vision to identify the boundaries of objects in an image. As such, I can see from the output image that it provides a clear demonstration of how edge detection works and how it can be used to highlight structural features like cracks. By implementing this code, I learned how this technique can help in tasks such as object detection, image segmentation, and feature extraction, where identifying boundaries or changes in intensity is important for further analysis.

## 3- Sobel Filter (Feature / Edge Detection)


```python
# Figure 3

from skimage import io
from skimage.filters import sobel
import matplotlib.pyplot as plt

# Get the image from GitHub
Image_url_3 = "https://raw.githubusercontent.com/Reem-KR/GEOL-558_Scikit-Image/main/Rock_Layers.png"

# Load image in grey scale
Image_3 = io.imread(Image_url_3, as_gray=True)

# Apply Sobel Filter
edges = sobel(Image_3)

# Create a figure with two subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Show original image
ax[0].imshow(Image_3, cmap='gray')
ax[0].set_title("Original Image of Rock Layers")
ax[0].axis("off")

# Show Sobel-Filtered image
ax[1].imshow(edges, cmap='gray')
ax[1].set_title("Sobel Filter Edge Detection of Rock Layers")
ax[1].axis("off")

# Display both images
plt.show()
```

**Figure 3:**

This image I took when I was on a geology field trip in Riyadh, Saudi Arabia looking at the layers of this formation. I chose this example to demonstrate the Sobel filter, which is an edge detection method. The Sobel filter detects edges by calculating the gradient of the image intensity at each pixel, which I think is very useful for highlighting boundaries in images. In this case, the rock layers have clear boundaries that can be accentuated using the Sobel filter. By implementing this example I saw how the Sobel Filter method is different in terms of output from the Canny Edge Detection method, since they are both considered edge detection techniques used to identify boundaries in images. The main difference I observed was how the Sobel Filter gave me thick and less precise edges and is more senstive to noise than the Canny Edge Detection. So, in the future I can use the Sobel Filter for quick edge detection that shows me the general idea then move on to the Canny Edge Detection for an accurate edge detection with noise reduction.

## 4- Histogram Equalization (Image Enhancement)



```python
# Figure 4.1

from skimage import io, exposure, color
import matplotlib.pyplot as plt

# Get the image from GitHub
Image_url_4 = "https://raw.githubusercontent.com/Reem-KR/GEOL-558_Scikit-Image/main/Mecca_From_Above.jpg"

# Load image in grey scale
Image_4 = io.imread(Image_url_4, as_gray=True)

# Apply Histogram Equalization to improve contrast
equalized = exposure.equalize_hist(Image_4)

# Create a figure with two subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Show original image
ax[0].imshow(Image_4, cmap='gray')
ax[0].set_title("Original Image of Mecca From Above")
ax[0].axis("off")

# Show Histogram Equalized image
ax[1].imshow(equalized, cmap='gray')
ax[1].set_title("Histogram Equalized Image of Mecca From Above")
ax[1].axis("off")

# Display both images
plt.show()
```

**Figure 4.1:**

This is a satellite image I found online of the city of Mecca. I chose this example to demonstrate Histogram Equalization, which is a technique that improves the contrast of an image by spreading out the intensity values. Here, I loaded the imgae in grayscale. This means that the image will have intensity values representing brightness, and there are no separate color channels (just a 2D array of pixel intensities). I learned that this process involves equalizing the histogram of the image, which means adjusting the pixel intensity values so that they are more evenly distributed across the available range. This can be very useful for enhancing images that appear dark or have poor contrast.

```python
# Figure 4.2

import numpy as np
import matplotlib.pyplot as plt
from skimage import io, exposure

# Get the image from GitHub
Image_url_4 = "https://raw.githubusercontent.com/Reem-KR/GEOL-558_Scikit-Image/main/Mecca_From_Above.jpg"

# Load image
Image_4 = io.imread(Image_url_4)

# Apply Histogram Equalization to each color channel separately
equalized_r = exposure.equalize_hist(Image_4[:, :, 0])  # Red channel
equalized_g = exposure.equalize_hist(Image_4[:, :, 1])  # Green channel
equalized_b = exposure.equalize_hist(Image_4[:, :, 2])  # Blue channel

# Stack the equalized channels back into a color image
equalized = np.stack([equalized_r, equalized_g, equalized_b], axis=-1)

# Create a figure with two subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Show original color image
ax[0].imshow(Image_4)
ax[0].set_title("Original Image of Mecca From Above")
ax[0].axis("off")

# Show Histogram Equalized color image
ax[1].imshow(equalized)
ax[1].set_title("Histogram Equalized Image of Mecca From Above")
ax[1].axis("off")

# Display both images
plt.show()
```

**Figure 4.2:**

This is the same image as in 4.1, a satellite image I found online of the city of Mecca. Here I also used the same Histogram Equalization technique but this time the image is loaded as a color image, and Histogram Equalization is applied to each channel (Red, Green, Blue) separately. I wanted to see the enhanced image contrast in color and how it differs from the greyscale. In general, I can see how Histogram Equalization is useful in image processing tasks where improved contrast is needed for better visibility or feature extraction, such as in medical imaging, satellite imaging, or computer vision applications.

## 5- Gaussian Blurring (Image Filtering / Smoothing)


```python
# Figure 5

from skimage import io
from skimage.filters import gaussian
import matplotlib.pyplot as plt

# Get the image from GitHub
Image_url_5 = "https://raw.githubusercontent.com/Reem-KR/GEOL-558_Scikit-Image/main/olivine.jpg"

# Load image
Image_5 = io.imread(Image_url_5)

# Apply Gaussian Blur
blurred = gaussian(Image_5, sigma=2)

# Create a figure with two subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Show original image
ax[0].imshow(Image_5, cmap='gray')
ax[0].set_title("Original Image of Olivine")
ax[0].axis("off")

# Show blurred image
ax[1].imshow(blurred, cmap='gray')
ax[1].set_title("Gaussian Blurred Image of Olivine")
ax[1].axis("off")

# Display both images
plt.show()
```

**Figure 5:**

This is an image I found online of an Olivine sample under the microscope.  I chose this example to demonstrate the Gaussian Blurring technique. This technique can be used to smooth images and reduce noise or details. This filter works by averaging pixel values within a neighborhood, weighted by a Gaussian distribution. As such, by implementing this, I learned how it can be useful to apply this method for noise reduction or feature smoothing before further analysis, such as Edge Detection.
